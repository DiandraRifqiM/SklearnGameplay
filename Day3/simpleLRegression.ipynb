Simple Linear Regression

#1 Datasets Sample

```python
import pandas as pd

pizza = {
    "rad":[6, 8, 10, 14, 18],
    "prices":[7, 9, 13, 17.5, 18]
}

pizzaDf = pd.DataFrame(pizza)
```

#2 Data Visualization

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")

pizzaDf.plot(kind='scatter', x="rad", y="prices")

plt.title("Pizza Data")
plt.xlabel("Radius")
plt.ylabel("Prices (USD)")
plt.xlim(0, 25)
plt.ylim(0, 25)
```

#3 Datasets Transform

```python
import numpy as np

X = np.array(pizzaDf["rad"])
y = np.array(pizzaDf["prices"])

print(f"X: {X}\nY: {y}")
```

```python
# Reshape X
X = X.reshape(-1, 1)
```

#4 Creat Simple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

linRegModel = LinearRegression()
linRegModel.fit(X, y)
```

#5 Previous Linear Regression Model Visualization

```python
X_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = linRegModel.predict(X_vis)
```

```python
plt.scatter(X, y, color='cyan', linewidths=0.5, edgecolors='blue')
plt.plot(X_vis, y_vis, '-r')

plt.title("Pizza Radius vs Prices Comparison")
plt.xlabel("Radius (inch)")
plt.ylabel("Prices (USD)")
plt.xlim(0, 25)
plt.ylim(0, 25)
```

Linear Regression Formual: y = α + βx

Note:

y = Response Variable

x = Explanatory Variable

α = Intercept

β = Slope

```python
print(f"Intercept: {linRegModel.intercept_}\nSlope:{linRegModel.coef_}")
```

#6 Pizza Price Predict

```python
pizzaDmtr = np.array([12, 17.3, 13.5]).reshape(-1, 1)
pizzaDmtr
```

```python
pricePred = linRegModel.predict(pizzaDmtr)
print(f"Price Predict: {pricePred}")
FixedPrice = np.round(pricePred, decimals=2)
print(f"Fixed Prices (USD): {FixedPrice}")
```

```python
for newDmtr, newPrice in zip(pizzaDmtr, FixedPrice):
    print(f"Pizza w/ diameter of {newDmtr}: ${newPrice}")
```

#7 Simple Linear Regression Model Evaluate

Training & Testing Dataset

```python
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = np.array([7, 9, 13, 17.5, 18])

X_test = np.array([8, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

Train Linear Regeression Model

```python
trainModel = LinearRegression()
trainModel.fit(X_train, y_train)
```

Linear Regeression Model Evaluate w/ Coef of Determination [R Squared]

```python
from sklearn.metrics import r2_score

y_pred = trainModel.predict(X_test)
rSquared = r2_score(y_test, y_pred)
print(f"R Squred: {rSquared}")
```
