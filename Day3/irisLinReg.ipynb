```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Sample Data

```python
pizza = {
    "Slice":[5, 7, 12, 16, 20],
    "Price":[3.55, 5.23, 7.35, 13.43, 16.44]
}

pizzaDf = pd.DataFrame(pizza)
```

Transform

```python
X = np.array(pizzaDf["Slice"])
y = np.array(pizzaDf["Price"])

# Reshape

X = X.reshape(-1, 1)
```

Simple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linRegModel = LinearRegression()
linRegModel.fit(X, y)
```

Data Visualization

```python
x_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = linRegModel.predict(x_vis)

plt.style.use("ggplot")
plt.scatter(X, y, color="cyan", edgecolors="blue", linewidths=0.5)
plt.plot(x_vis, y_vis, '-r')
plt.title("Pizza Slice vs Price Comparison")
plt.xlabel("Pizza Slices")
plt.ylabel("Prices (USD)")
plt.xlim(0, 25)
plt.ylim(0, 25)
```

Pizza Price Predict

```python
newSlice = np.array([14, 6, 18, 25]).reshape(-1, 1)
predPrice = linRegModel.predict(newSlice)
fixedPrice = np.round(predPrice, decimals=2)
```

```python
for slices, price in zip(newSlice, fixedPrice):
    print(f"Pizza Slices:{slices} : ${price}")
```
