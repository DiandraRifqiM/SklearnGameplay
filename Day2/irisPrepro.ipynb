Dataset Prepreation

```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Data Splitting

iris = load_iris()
X = iris.data
y = iris.target
```

Binarization

```python
from sklearn import preprocessing

preprocess = preprocessing.Binarizer(threshold=0.5)
binerizedDataX = preprocess.transform(X)
```

Scalling

```python
preprocess = preprocessing.MinMaxScaler(feature_range=(0, 1))
scaleData = preprocess.fit_transform(X)
```

Least Absolut Deviations

```python
normDataL1 = preprocessing.normalize(X, norm='l1')
normDataL1

import matplotlib.pyplot as plt
plt.style.use("ggplot")

X1 = normDataL1[:, 0]
X2 = normDataL1[:, 1]

Xa = X[:, :2]

plt.scatter(X1, X2, color="cyan", linewidths=0.5, edgecolors="blue")

m, b = np.polyfit(X1, X2, 1)

yFit = m * X1 + b

plt.plot(X1, yFit, color="r", label=f"Least Squares Fit: y = {m:.2f}x + {b:.2f}")

plt.xlabel('Normalized Feature 0')
plt.ylabel('Normalized Feature 1')
plt.title('L1 Normalized Iris Data (First Two Features)')
```
