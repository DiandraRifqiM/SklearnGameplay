Decision Tree Classification

Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
X.shape
```

Train Test Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Train Model

```python
from sklearn.tree import DecisionTreeClassifier

deciTree = DecisionTreeClassifier(max_depth=3)
deciTree.fit(X_train, y_train)
```

Decision Tree Visualize

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 100
fig, ax = plt.subplots(figsize=(8, 6))
tree.plot_tree(deciTree, fontsize=10, ax=ax, max_depth=3, proportion=True, precision=2)
plt.title("Iris Decision Tree", fontsize=14)
plt.tight_layout()
```

Evaluate Model

```python
from sklearn.metrics import classification_report

yPred = deciTree.predict(X_test)
report = classification_report(y_test, yPred)
print(f"Decision Tree Prediction Report:\n{report}")
```
