Dataset

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
```

Data Splitting

```python
lbc = load_breast_cancer()
X = lbc.data
y = lbc.target
```

Preprocessing

```python
from sklearn.preprocessing import PolynomialFeatures
quadrat = PolynomialFeatures(degree=2)
X_scaled = quadrat.fit_transform(X)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)
```

Model Training

```python
from sklearn.neighbors import KNeighborsRegressor
K = 4
knModel = KNeighborsRegressor(n_neighbors=K)
knModel.fit(X_train, y_train)
```

Model Evaluate

```python
from sklearn.metrics import r2_score as r2
yPred = knModel.predict(X_test)

r2Score = r2(y_test, yPred)
print(f"KNN Regressor Prediciton Score: {r2Score}")
```

Data Visualization

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# decomposition
pca = PCA(n_components=2)
X_train2d = pca.fit_transform(X_train)

plt.style.use('ggplot')
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_train2d[:, 0], X_train[:, 1], c=y_train, 
                      cmap="cool", edgecolor="k", linewidths=0.5)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("2D Scatter Plot (PolynomialFeatures, No Scaling)")
```
