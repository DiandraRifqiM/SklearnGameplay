Datasets Preparation

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, yTrain, yTest = train_test_split(X, y, random_state=1, test_size=0.4)
```

Model Training

```python
from sklearn.neighbors import KNeighborsClassifier

knModel = KNeighborsClassifier(n_neighbors=5)
knModel.fit(X_train, yTrain)
```

Model Evaluate

```python
from sklearn.metrics import accuracy_score

yPred = knModel.predict(X_test)
accuracy = accuracy_score(yTest, yPred)
print(f"Accuracy: {accuracy}")
```

Model Applied w/ New Dataset

```python
newData = [[3, 4, 2, 2],
           [5, 2, 1, 2]]

pred = knModel.predict(newData)

predSpecies = [iris.target_names[p] for p in pred]
print(f"Prediction Result: {predSpecies}")
```

Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")

targetName = iris.target_names
cmap = plt.get_cmap("cool")
colors = cmap(np.linspace(0, 1, len(targetName)))


for i, species in enumerate(targetName):
    plt.scatter(
        X[y == i, 0],
        X[y == i, 1],
        label = species,
        color = colors[i]
    )

xMin, xMax = X[:, 0].min() - 0.5, X[:, 0].max()-0.5
yMin, yMax = X[:, 1].min()-0.5, X[:, 1].max()-0.5

plt.xlim(xMin, xMax)
plt.ylim(yMin, yMax)
plt.title("Sepal Length vs Sepal Width of Iris")
plt.xlabel("Sepal Length", labelpad=20)
plt.ylabel("Sepal Width", labelpad=20)
plt.legend(title="Species", bbox_to_anchor=(1.3, 0.95))
```
