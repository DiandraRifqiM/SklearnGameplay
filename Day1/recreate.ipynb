Data Preperation

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split

XTrain, XTest, yTrain, yTest = train_test_split(X, y, train_size=0.1, random_state=1)
```

Model Training

```python
from sklearn.neighbors import KNeighborsClassifier

KNmodel = KNeighborsClassifier(n_neighbors= 3)
KNmodel.fit(XTrain, yTrain)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score

yPred = KNmodel.predict(XTest)
acc = accuracy_score(yTest, yPred)

print(f"Accuracy: {acc}")
```

Model Training Applied

```python
newData = [[3, 1, 5, 5],
           [4, 1, 1, 2]]

testPred = KNmodel.predict(newData)
testPred
```

```python
result = [iris.target_names[p] for p in testPred]
print(f"Hasil Prediksi: {result}")
```

Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")

targetName = iris.target_names
cmap = plt.get_cmap("cool")
colors = cmap(np.linspace(0, 1, len(targetName)))


for i, species in enumerate(targetName):
    plt.scatter(
        X[y == i, 0],
        X[y == i, 2],
        label = species,
        color = colors[i]
    )


xMin, xMax = X[:, 0].min() - 0.5, X[:, 0].max() - 0.5
yMin, yMax = X[:, 2].min() - 0.5, X[:, 2].max() - 0.5

# plt.scatter(X[:, 0], X[:, 2], c=y, cmap="cool")
plt.title("Iris Sepal vs Petal (Length)")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.xlim(xMin, xMax)
plt.ylim(yMin, yMax)
plt.legend(title="Species", bbox_to_anchor=(1.3, 0.95))
```
